````python
def get_unigram_counts(corpus):
    """
    Given a corpus, calculates the unigram counts for each character in the corpus

    Args:
        corpus [list[list[str]]]: list of tokenized characters. Text is appended with END token.

    Returns:
        unigram_counts [dict [key: char, value: count]]:
            dictionary of unigram counts for each character in the corpus
        Example:
        > unigram_counts["c1"] = 5
    """

    # ADD YOUR CODE HERE
    # BEGIN CODE
    unigram_counts = defaultdict(int)
    for name in corpus:
        for char in name[:]:
            unigram_counts[char] += 1
    # END CODE
    return unigram_counts
````

````python
"""
Implementaion of a Unigram Model without smoothing
"""

class UnigramModel(NGramLanguageModel):
    def __init__(self, train_text):
        """
        Initialise and train the model with train_text.

        Args:
            train_text [list of list]: list of tokenised names
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        super().__init__(train_text)

        self.unigram_counts = get_unigram_counts(train_text)
        self.vocab_size = vocab_size
        # END CODE


    def get_next_char_probabilities(self):
        """
        Return a dictionary of probabilities for each char in the vocabulary

        Returns:
            key: char, value: probability
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        total_count = sum(self.unigram_counts.values())
        next_char_probabilities = {char: count / total_count for char, count in self.unigram_counts.items()}
        return next_char_probabilities
        # END CODE





    def get_name_log_probability(self, name):
        """
        Calculates the log probability of name according to the n-gram model

        Args:
            name [list]: list of tokens

        Returns:
            Log probability of the name [float]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        log_prob = sum(math.log(self.get_next_char_probabilities().get(char, 1/self.vocab_size)) for char in name)
        return log_prob
        # END CODE





    def get_perplexity(self, text):
        """
        Returns the perplexity of the model on a text as a float.

        Args:
            text [list]: a list of string tokens

        Returns:
            perplexity of the given text [float]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        total_log_prob = 0.0
        total_chars = 0.0
        for name in text:
            prob = self.get_name_log_probability(name)
            total_log_prob += prob
            for char in name:
                total_chars += 1


        avg_log_prob = total_log_prob / total_chars
        perplexity = math.exp(-avg_log_prob)
        return perplexity
        # END CODE




    def generate_names(self, k, n=MAX_NAME_LENGTH, prefix=None):
        """
        Given a prefix, generate k names according to the model.
        The default prefix is None.

        Args:
            k [int]: Number of names to generate
            n [int]: Maximum length (number of tokens) in the generated name
            prefix [list of tokens]: Prefix after which the names have to be generated

        Returns:
            list of generated names [list]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        names = []
        for i in range(k):
            name = prefix.copy() if prefix else []
            while len(name) < n:
                next_char_probs = self.get_next_char_probabilities()
                char = random.choices(list(next_char_probs.keys()), weights=list(next_char_probs.values()))[0]
                if char ==START:
                    continue
                if char == END:
                    break
                name.append(char)
            names.append(''.join(name))
        return names
        # END CODE






    def get_most_likely_chars(self, sequence, k):
        """
        Given a sequence of characters, outputs k most likely characters after the sequence.

        Args:
            sequence [list[str]]: list of characters
            k [int]: number of most likely characters to return

        Returns:
            chars [list[str]]: *Ordered* list of most likely characters
                        (with charcater at index 0 being the most likely and
                        character at index k-1 being the least likely)

        """
        # ADD YOUR CODE HERE
        # BEGIN CODE
        next_char_probs = self.get_next_char_probabilities()
        most_likely_chars = sorted(next_char_probs.keys(), key=lambda x: -next_char_probs[x])[:k]
        return most_likely_chars
        # END CODE

````

````python
"""
Implementation of a Bigram Model.
"""

class BigramModel(NGramLanguageModel):
    def __init__(self, train_text):
        """
        Initialise and train the model with train_text.

        Args:
            train_text [list of list]: list of tokenised names
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        self.bigram_counts = get_bigram_counts(train_text)
        self.vocab_size = vocab_size
        # END CODE

    def get_next_char_probabilities(self):
        """
        Returns a probability distribution over all chars in the vocabulary.
        Probability distribution should sum to one.

        Returns:
            P: dictionary or nested dictionary; Output format depends on n-gram
            Examples:
                for N=2 (bigram); dict[key:bigram_char1, value:dict[key:bigram_char2,value:probability of bigram]]
                    > P["a"]["b"] = 0.0001 (which stands of P("b"|"a"))
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        next_char_probabilities = {}
        for char1 in self.bigram_counts:
            total_count = sum(self.bigram_counts[char1].values())
            next_char_probabilities[char1] = {char2: count / total_count for char2, count in self.bigram_counts[char1].items()}
        return next_char_probabilities
        # END CODE






    def get_name_log_probability(self, name):
        """
        Calculates the log probability of name according to the n-gram model.
        Be careful with cases for which probability of the name is zero.

        Args:
            name [list]: list of tokens

        Returns:
            Log probability of the name [float]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        log_probability = 0.0
        prior_char = START
        for char in name:
            if char in self.bigram_counts[prior_char]:
                log_probability += math.log(self.bigram_counts[prior_char][char] / sum(self.bigram_counts[prior_char].values()))
            else:
                log_probability += math.log(1 / self.vocab_size)
            prior_char = char
        return log_probability
        # END CODE



    def get_perplexity(self, text):
        """
        Returns the perplexity of the model on a text as a float.

        Args:
            text [list]: a list of string tokens

        Returns:
            perplexity of the given text [float]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        log_likelihood = sum([self.get_name_log_probability(name) for name in text])
        perplexity = math.exp(-log_likelihood / len([char for name in text for char in name]))
        return perplexity
        # END CODE





    def generate_names(self, k, n=MAX_NAME_LENGTH, prefix=None):
        """
        Given a prefix, generate k names according to the model.
        The default prefix is None.

        Args:
            k [int]: Number of names to generate
            n [int]: Maximum length (number of tokens) in the generated name
            prefix [list of tokens]: Prefix after which the names have to be generated

        Returns:
            list of generated names [list]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        generated_names = []
        for _ in range(k):
          name = prefix.copy() if prefix else []
          prior_char = name[-1] if name else START
          for _ in range(n):
            if prior_char in self.bigram_counts:
                next_char_probabilities = self.bigram_counts[prior_char]
                next_char = random.choices(list(next_char_probabilities.keys()), list(next_char_probabilities.values()))[0]
                if next_char ==START:
                    continue
                if next_char == END:
                    break
                name.append(next_char)
                prior_char = next_char
            else:
                break
          generated_names.append("".join(name))
        return generated_names
        # END CODE


    def get_most_likely_chars(self, sequence, k):
        """
        Given a sequence of characters, outputs k most likely characters after the sequence.

        Args:
            sequence [list[str]]: list of characters
            k [int]: number of most likely characters to return

        Returns:
            chars [list[str]]: *Ordered* list of most likely characters
                        (with charcater at index 0 being the most likely and
                        character at index k-1 being the least likely)

        """
        # ADD YOUR CODE HERE
        # BEGIN CODE
        if len(sequence) > 0:
            prior_char = sequence[-1]
        else:
            prior_char = START
        if prior_char in self.bigram_counts:
            sorted_chars = sorted(self.bigram_counts[prior_char].items(), key=lambda x: x[1], reverse=True)
            return [char for char, _ in sorted_chars[:k]]
        return []
        # END CODE

````

````python
"""
Implementation of a bigram model with laplace or add-k smoothing.

"""

class LaplaceSmoothedBigramModel(BigramModel):
    # This class extends BigramModel.

    def __init__(self, train_text, k):
        super().__init__(train_text)
        self.k = k # specify k for smoothing

    # You should override ONLY those functions
    # which calculate probability of a bigram.
    # You can override get_next_char_probabilities
    # or any other helper functions you use in BigramModel
    # to calculate bigram probabilities.
    # BEGIN CODE
    def get_smoothed_probabilities(self):
        smoothed_probs = {}
        for char1 in self.bigram_counts:
            total_count = sum(self.bigram_counts[char1].values()) + (self.vocab_size * self.k)
            probs = {}
            for char2, count in self.bigram_counts[char1].items():
                probs[char2] = (count + self.k) / total_count
            # Smoothing for unseen bigrams
            for char in self.bigram_counts:
                if char not in probs:
                    probs[char] = self.k / total_count
            # Normalize probabilities to sum to 1
            total_prob = sum(probs.values())
            smoothed_probs[char1] = {char: prob / total_prob for char, prob in probs.items()}
        return smoothed_probs

    def get_next_char_probabilities(self):
        return self.get_smoothed_probabilities()
    # END CODE
````

#### FNN Implementation

**Feed-forward Neural Language Modelling**

Like the n-gram LM, the feedforward neural LM approximates the probability of a char given the entire prior context $P(w_t|w_{1:t−1})$ by approximating based on the $N-1$ previous chars:
$$P(w_t|w_1,...,w_{t−1}) ≈ P(w_t|w_{t−N+1},...,w_{t−1})$$


Implement the FNN LM given in this paper: [Neural Probabilistic Language Model](https://www.jmlr.org/papers/volume3/bengio03a/bengio03a.pdf)

The architecture of the FNN can be described by the equation and figure:

$$y = b + W x + U \text t \text a \text n \text h (d + H x)$$

- $x$ is of size $(ngram-1)*m$ where $m$ is the size embedding dimensions
- $y$ is of size $V*1$ where $V$ is the vocabulary size

![FNN_LM](https://drive.google.com/uc?id=1aQhkXjWelHfiBfmBQV3z5TjHFNMtqtzT)


**Some tips**:
- embed the chars with dimension $m$ (example, $60$), then flatten into a single embedding for  $n-1$  chars (with size  $(n-1)*m$ )
- you can use Adam or Stochastic Gradient Descent (SGD) for optimising the cross entropy loss
- If you are using SGD, you may want to use momentum, and a learning rate scheduler
- do early stopping based on validation set loss or perplexity

**Important**: Fix seed as 42 whenever performing any randomized operations, e.g., initializing ML models.

````python
"""
Implemenation of a PyTorch Module that holds the neural network for your model

"""
class FNN_LM(nn.Module):

    def __init__(self, vocab_size, emb_size, hid_size, ngram):
        super(FNN_LM, self).__init__()
        self.ngram = ngram

        # YOUR CODE HERE
        # BEGIN CODE
        self.embedding = nn.Embedding(vocab_size, emb_size)
        self.linear1 = nn.Linear((ngram-1) * emb_size, hid_size)
        self.linear2 = nn.Linear(hid_size, vocab_size)
        self.tanh = nn.Tanh()
        # END CODE






    def forward(self, chars):
        """
        Args:
            chars: this is a tensor of inputs with shape [batch_size x ngram-1]

        Returns:
            logits: a tensor of log probabilities with shape [batch_size x vocab_size]

        """
        # Embedding input characters
        x = self.embedding(chars)

        # Flatten the embedding
        x = x.view(-1, (self.ngram-1) * self.embedding.embedding_dim)

        # Pass the input through the first linear layer and apply the tanh activation function
        x = self.linear1(x)
        x = self.tanh(x)

        # Pass the output of the first linear layer through the second linear layer
        logits = self.linear2(x)
        # END CODE

        return logits

````

For this stage of the assignment, you will implement an RNN language model.

Some tips:
* use dropout
* use the same weights for the embedding layer and the pre-softmax layer

````python
"""
Implemenation of a PyTorch Module that holds the RNN

"""
class RNN_LM(nn.Module):
  def __init__(self, vocab_size, embedding_dim, hidden_dim, dropout):
        super(RNN_LM, self).__init__()
        # BEGIN CODE

        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, batch_first=True)
        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim, vocab_size)

        # Initialize weights for the embedding and pre-softmax layer
        nn.init.xavier_uniform_(self.embedding.weight)
        nn.init.xavier_uniform_(self.fc.weight)

        # Use the same weights for the embedding and pre-softmax layer
        self.fc.weight = self.embedding.weight

  def forward(self, x, hidden):
        embedded = self.embedding(x)
        output, hidden = self.rnn(embedded, hidden)
        output = self.dropout(output)
        logits = self.fc(output)
        return logits, hidden
        # END CODE
````

````python
class RNNTrainer:
    """
    RNNTrainer wraps RNN_LM to handle training and evaluation.

    """

    # NOTE: you are free to add additional inputs/functions
    # to RNNTrainer to make training better
    # make sure to define and add it within the input
    # and initialization if you are using any additional inputs
    # for usage in the function

    def __init__(
        self,
        model,
        optimizer,
        criterion,
        train_dataloader,
        valid_dataloader,
        epochs,
        use_cuda,
        vocab,
        model_dir
    ):

        self.model = model
        self.epochs = epochs
        self.optimizer = optimizer
        self.criterion = criterion
        self.train_dataloader = train_dataloader
        self.valid_dataloader = valid_dataloader
        self.use_cuda = use_cuda
        self.model_dir = model_dir
        self.loss = {"train": [], "val": []}
        self.vocab = vocab

        # Move the model to GPU if available
        if self.use_cuda:
            self.model = self.model.cuda()

    def train(self):
        """
        Trains the model with train_dataloader and validates using valid_dataloader

        """
        # You may change the input arguments to this function,
        # but make sure to also change the code wherever this function is called

        # ADD YOUR CODE HERE FOR TRAINING & VALIDATION
        # BEGIN CODE

        for epoch in range(self.epochs):
            self.model.train()
            total_loss = 0.0
            for batch_input, batch_output in tqdm(self.train_dataloader):
                hidden = None
                if self.use_cuda:
                    batch_input, batch_output = batch_input.cuda(), batch_output.cuda()
                self.optimizer.zero_grad()
                output, _ = self.model(batch_input, hidden)
                loss = self.criterion(output.view(-1, len(self.vocab)), batch_output.view(-1))
                loss.backward()
                self.optimizer.step()
                total_loss += loss.item()
            train_loss = total_loss / len(self.train_dataloader)
            self.loss["train"].append(train_loss)
            # Validation
            self.model.eval()
            with torch.no_grad():
                val_loss = 0
                for batch_input, batch_output in self.valid_dataloader:
                    hidden = None
                    if self.use_cuda:
                        batch_input, batch_output = batch_input.cuda(), batch_output.cuda()
                    output, _ = self.model(batch_input, hidden)
                    loss = self.criterion(output.view(-1, len(self.vocab)), batch_output.view(-1))
                    val_loss += loss.item()
                val_loss /= len(self.valid_dataloader)
                self.loss["val"].append(val_loss)
            print(f"Epoch {epoch+1}/{self.epochs}, Train Loss: {train_loss}, Val Loss: {val_loss}")
            # END CODE





    def save_model(self):
        """
        Save final model to directory

        """

        model_path = os.path.join(self.model_dir, "model.pt")
        torch.save(self.model, model_path)


    def save_loss(self):
        """
        Save train/val loss as json file to the directory

        """

        loss_path = os.path.join(self.model_dir, "loss.json")
        with open(loss_path, "w") as fp:
            json.dump(self.loss, fp)


    def get_next_char_probabilities(self,sequence=None):
        """
        Return a dictionary of probabilities for each char in the vocabulary
        with a default starting sequence of [START]

        Returns:
            dictionary with key: char, value: probability

        """

        # ADD YOUR CODE HERE
        # BEGIN CODE
        self.model.eval()
        stoi = self.vocab.get_stoi()
        itos = self.vocab.get_itos()
        if sequence is None:
            sequence = [START]
        prefix_ids = [self.vocab[token] for token in sequence]
        with torch.no_grad():
            hidden = None
            for token_id in prefix_ids:
                output, hidden = self.model(torch.tensor([[token_id]]).cuda(), hidden)
            probabilities = F.softmax(output.squeeze(), dim=0).cpu().numpy()
        char_probabilities = {itos[i]: probabilities[i] for i in range(len(probabilities))}
        return char_probabilities
        # END CODE



    def generate_names(self, k, n, prefix=None):
        """
        Given a prefix, generate k names according to the model.
        The default prefix is None.

        Args:
            k [int]: Number of names to generate
            n [int]: Maximum length (number of tokens) in the generated name
            prefix [list of tokens]: Prefix after which the names have to be generated

        Returns:
            list of generated names [list[str]]
        """

        # ADD YOUR CODE HERE
        # BEGIN CODE

        # don't forget self.model.eval()
        self.model.eval()
        stoi = self.vocab.get_stoi()
        itos = self.vocab.get_itos()
        if prefix is None:
            prefix = [START]
        names = []
        with torch.no_grad():
            for i in range(k):
                generated_name = prefix.copy()
                hidden = None
                for j in range(n):
                    input_tensor = torch.tensor([[self.vocab[token] for token in generated_name[-1:]]]).cuda()
                    output, hidden = self.model(input_tensor, hidden)
                    probabilities = F.softmax(output.squeeze(), dim=0).cpu().numpy()
                    next_token_id = np.random.choice(len(probabilities), p=probabilities)
                    next_token = itos[next_token_id]
                    if next_token ==START:
                        continue
                    if next_token == END:
                        break
                    generated_name.append(next_token)
                names.append("".join(generated_name[:]))  # Exclude the START token
        return names
        # END CODE





    def get_perplexity(self, text):
        """
        Returns the perplexity of the model on text as a float.

        Args:
            text [list[list[str]]]: list of tokenised names
            > Example:
            [['<s>', 'a', 'a', 'b', 'i', 'd', '</s>'],
            ['<s>', 'a', 'a', 'b', 'i', 'd', 'a', '</s>']]

        Returns:
            perplexity [float]

        """

        # ADD YOUR CODE HERE
        # BEGIN CODE

        # you may want to use the dataloader here
        # don't forget self.model.eval()
        self.model.eval()
        with torch.no_grad():
            total_loss = 0
            total_tokens = 0
            for batch_input, batch_output in self.valid_dataloader:
                hidden = None
                if self.use_cuda:
                    batch_input, batch_output = batch_input.cuda(), batch_output.cuda()
                output, _ = self.model(batch_input, hidden)
                loss = self.criterion(output.view(-1, len(self.vocab)), batch_output.view(-1))
                total_loss += loss.item() * batch_input.size(0)
                total_tokens += batch_input.size(0) * batch_input.size(1)
            perplexity = math.exp(total_loss / total_tokens)
        return perplexity
        # END CODE




    def get_most_likely_chars(self, sequence, k):
        """
        Given a sequence of characters, outputs k most likely characters after the sequence.

        Args:
            sequence [list[str]]: list of characters
            k [int]: number of most likely characters to return

        Returns:
            chars [list[str]]: *Ordered* list of most likely characters
                        (with charcater at index 0 being the most likely and
                        character at index k-1 being the least likely)

        """

        # ADD YOUR CODE HERE
        # BEGIN CODE

        # don't forget self.model.eval()
        self.model.eval()
        stoi = self.vocab.get_stoi()
        itos = self.vocab.get_itos()
        with torch.no_grad():
            hidden = None
            for token in sequence:
                input_tensor = torch.tensor([[self.vocab[token]]]).cuda()
                output, hidden = self.model(input_tensor, hidden)
            probabilities = F.softmax(output.squeeze(), dim=0).cpu().numpy()
            most_likely_ids = np.argsort(probabilities)[::-1][:k]
            most_likely_chars = [itos[i] for i in most_likely_ids]
        return most_likely_chars
        # END CODE


````
